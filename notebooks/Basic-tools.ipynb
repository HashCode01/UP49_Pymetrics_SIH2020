```python
def prepare_set(df,i):
    df['pid'] = i
    df['time'] = df['Unnamed: 0']
    df = df.drop(columns = ['Unnamed: 0'])
    return df
```

```python
min_vals = {'Temp' :34 ,'BaseExcess': -3,'HCO3':20 ,'FiO2':0.0 ,'pH':7.3,'PaCO2': 28,'SaO2': 0,'AST': 0,'BUN': 5,'Alkalinephos':10 ,'Calcium': 7,'Chloride': 95,'Creatinine': 0.4,'Bilirubin_direct': 0,'Glucose': 50,'Magnesium': 1.2, 'Potassium': 3.0, 'Hct':31 , 'Hgb':6.5 , 'PTT':20 , 'WBC': 2, 'Platelets': 100}
max_vals = {'Temp' :42 ,'BaseExcess': 3,'HCO3':30 ,'FiO2':1.0 ,'pH': 7.5,'PaCO2': 48 ,'SaO2': 100,'AST': 40,'BUN': 25,'Alkalinephos':135 ,'Calcium': 12.5,'Chloride': 110,'Creatinine': 1.6,'Bilirubin_direct': 0.6,'Glucose': 120,'Magnesium': 2.8, 'Potassium': 6.0, 'Hct':56 , 'Hgb':19 , 'PTT':40 , 'WBC': 15, 'Platelets': 400}
def imputation(X,Y,i):
    X_arr = pd.DataFrame(X).to_numpy(dtype=float)
    Y_arr = pd.DataFrame(Y).to_numpy(dtype=float)
    
    
    add = np.concatenate((X_arr,Y_arr),axis=1)
    imp = IterativeImputer(max_iter=10, random_state=0,sample_posterior=True,min_value = min_vals[i],max_value = max_vals[i])
    imp.fit(add)
    impute = np.round(imp.transform(add),decimals=3)
    
    return impute
```

```python
def prepare_DS(df):
    df['HR']=df['HR'].fillna(df['HR'].median())
    df['O2Sat']=df['O2Sat'].fillna(df['O2Sat'].median())
    
    impute_list_1 = ['Temp']
    for i in impute_list_1:
        df[i] = pd.DataFrame(imputation(df[i],df['SepsisLabel'],i),columns=['base_imp','SepsisLabel'])['base_imp']
        
    df['SBP']=df['SBP'].fillna(df['SBP'].median())
    df['MAP']=df['MAP'].fillna(df['MAP'].median())
    df['DBP']=df['DBP'].fillna(df['DBP'].median())
    df['Resp']=df['Resp'].fillna(df['Resp'].median())
    
    df = df.drop(columns = ['EtCO2'])
    
    impute_list_2 = ['BaseExcess', 'HCO3', 'FiO2','pH','PaCO2','SaO2','AST','BUN','Alkalinephos','Calcium','Chloride','Creatinine','Bilirubin_direct','Glucose']
    for i in impute_list_2:
        df[i] = pd.DataFrame(imputation(df[i],df['SepsisLabel'],i),columns=['base_imp','SepsisLabel'])['base_imp']
        
    df = df.drop(columns = ['Lactate'])
    
    impute_list_3 = ['Magnesium']
    for i in impute_list_3:
        df[i] = pd.DataFrame(imputation(df[i],df['SepsisLabel'],i),columns=['base_imp','SepsisLabel'])['base_imp']
        
    df = df.drop(columns = ['Phosphate'])
    
    impute_list_4 = ['Potassium']
    for i in impute_list_4:
        df[i] = pd.DataFrame(imputation(df[i],df['SepsisLabel'],i),columns=['base_imp','SepsisLabel'])['base_imp'] 
        
    df = df.drop(columns = ['Bilirubin_total', 'TroponinI'])
    
    impute_list_5 = ['Hct', 'Hgb', 'PTT', 'WBC']
    for i in impute_list_5:
        df[i] = pd.DataFrame(imputation(df[i],df['SepsisLabel'],i),columns=['base_imp','SepsisLabel'])['base_imp']
        
    df = df.drop(columns = ['Fibrinogen'])
    
    impute_list_6 = ['Platelets']
    for i in impute_list_6:
        df[i] = pd.DataFrame(imputation(df[i],df['SepsisLabel'],i),columns=['base_imp','SepsisLabel'])['base_imp']
        
    df['Unit1']=df['Unit1'].fillna(1)
    df['Unit2'] = df['Platelets']
    df['HospAdmTime']=df['HospAdmTime'].fillna(df['HospAdmTime'].mean())
    
    df.head()
    return df
```

### SOFA Score Calculations

```python
def Sofa_Calc(df):
    for i in df.index:
        df['paO2_FiO2']= 0
        df['Sofa_Renal']= 0
        df['Sofa_Resp'] = 0
        df['Sofa_Coag'] = 0
        df['Sofa_Score'] = 0
        #Respiration
        df['paO2_FiO2'][i] = 713 - (df['PaCO2'][i])/(0.8 * df['FiO2'][i])
        if(df['paO2_FiO2'][i] < 100):
            df['Sofa_Resp'][i] = 4
        elif(df['paO2_FiO2'][i] < 200):
            df['Sofa_Resp'][i] = 3
        elif(df['paO2_FiO2'][i] < 300):
            df['Sofa_Resp'][i] = 2
        elif(df['paO2_FiO2'][i] < 400):
            df['Sofa_Resp'][i] = 1
        else:
            df['Sofa_Resp'][i] = 0
        
        #Renal
        if(df['Creatinine'][i] >= 5):
            df['Sofa_Renal'][i] = 4
        elif(df['Creatinine'][i] >=3.5):
            df['Sofa_Renal'][i] = 3
        elif(df['Creatinine'][i] >=2.0):
            df['Sofa_Renal'][i] = 2
        elif(df['Creatinine'][i] >=1.2):
            df['Sofa_Renal'][i] = 1
        else:
            df['Sofa_Renal'][i] = 0
    
        #Coagulation
        if(df['Platelets'][i] < 20):
            df['Sofa_Coag'][i] = 4
        elif(df['Platelets'][i] < 50):
            df['Sofa_Coag'][i] = 3
        elif(df['Platelets'][i] < 100):
            df['Sofa_Coag'][i] = 2
        elif(df['Platelets'][i] < 150):
            df['Sofa_Coag'][i] = 1
        else:
            df['Sofa_Coag'][i] = 0
        
        df['Sofa_Score'][i] = df['Sofa_Coag'][i] + df['Sofa_Resp'][i] + df['Sofa_Renal'][i]
    return df
```
