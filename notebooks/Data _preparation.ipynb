```python
import pandas as pd
import numpy as np

def to_str(i):
    st = str(i)
    return (5-len(st))*'0' + st

lst = ['Age', 'ICULOS', 'HAT', 'Gender', 'Temp', 'qSOFA', 'PaO2/FiO2', 'SOFA_score', 'sepsis_label']
Data = pd.DataFrame(columns = lst)

for i in range(1,18001):
    filename = "training_A\p0"+to_str(i)+".csv"
    d1 = pd.read_csv(filename)
    
    output = dict()

    output['Age'] = d1['Age'][0]

    output['ICULOS'] = d1['ICULOS'][0]

    output['HR'] = d1['HR'].mean()

    output['HAT'] = d1['HospAdmTime'][0]

    output['Gender'] = d1['Gender'][0]
    
    if d1['Temp'].min()<36 or d1['Temp'].max()>38:
        output['Temp'] = 1 
    else :
        if d1['Temp'].min()!=np.NaN : 
            output['Temp'] = 0
        else: output['Temp'] = np.NaN

    if d1['Resp'].max()>22 or d1['SBP'].min()<100:
        output['qSOFA'] = 1
    else:  output['qSOFA'] = 0
    
    if d1['PaCO2'].mean()!=np.NaN and d1['FiO2'].mean()!=np.NaN :
        output['PaO2/FiO2'] = 713-((5/4)*(d1['PaCO2'].mean()/d1['FiO2'].mean()))
    else: output['PaO2/FiO2'] = np.NaN
        
    Platelets = d1['Platelets'].min()
    Bilirubin = d1['Bilirubin_total'].max()
    Creatinine = d1['Creatinine'].max()

    MAP = d1['MAP'].min()
    SOFA_score = 0
    if MAP<70:
        SOFA_score+=1

    if output['PaO2/FiO2']<100:
        SOFA_score += 1
    if output['PaO2/FiO2']<200:
        SOFA_score += 1
    if output['PaO2/FiO2']<300:
        SOFA_score += 1
    if output['PaO2/FiO2']<400:
        SOFA_score += 1

    if Bilirubin>1.2:
        SOFA_score+=1
    if Bilirubin>2:
        SOFA_score+=1
    if Bilirubin>6:
        SOFA_score+=1
    if Bilirubin>12:
        SOFA_score+=1


    if Platelets<20:
        SOFA_score+=1
    if Platelets<50:
        SOFA_score+=1
    if Platelets<100:
        SOFA_score+=1
    if Platelets<150:
        SOFA_score+=1

    if Creatinine>1.2:
        SOFA_score+=1
    if Creatinine>2:
        SOFA_score+=1
    if Creatinine>3.5:
        SOFA_score+=1
    if Creatinine>5:
        SOFA_score+=1
        
    output['SOFA_score']=SOFA_score
# old
#      if MAP!=np.NaN and MAP<70:
#         SOFA_score+=1

#     if output['PaO2/FiO2']!=np.NaN and output['PaO2/FiO2']<100:
#         SOFA_score += 1
#     if output['PaO2/FiO2']!=np.NaN and output['PaO2/FiO2']<200:
#         SOFA_score += 1
#     if output['PaO2/FiO2']!=np.NaN and output['PaO2/FiO2']<300:
#         SOFA_score += 1
#     if output['PaO2/FiO2']!=np.NaN and output['PaO2/FiO2']<400:
#         SOFA_score += 1

#     if Bilirubin!=np.NaN and Bilirubin>1.2:
#         SOFA_score+=1
#     if Bilirubin!=np.NaN and Bilirubin>2:
#         SOFA_score+=1
#     if Bilirubin!=np.NaN and Bilirubin>6:
#         SOFA_score+=1
#     if Bilirubin!=np.NaN and Bilirubin>12:
#         SOFA_score+=1


#     if Platelets!=np.NaN and Platelets<20:
#         SOFA_score+=1
#     if Platelets!=np.NaN and Platelets<50:
#         SOFA_score+=1
#     if Platelets!=np.NaN and Platelets<100:
#         SOFA_score+=1
#     if Platelets!=np.NaN and Platelets<150:
#         SOFA_score+=1

#     if Creatinine!=np.NaN and Creatinine>1.2:
#         SOFA_score+=1
#     if Creatinine!=np.NaN and Creatinine>2:
#         SOFA_score+=1
#     if Creatinine!=np.NaN and Creatinine>3.5:
#         SOFA_score+=1
#     if Creatinine!=np.NaN and Creatinine>5:
#         SOFA_score+=1

   
    if d1['SepsisLabel'].mean()>0.01:
        output['sepsis_label'] = 1
    else:
        output['sepsis_label'] = 0

    Data = Data.append(output, ignore_index = True)
    
Data
```
