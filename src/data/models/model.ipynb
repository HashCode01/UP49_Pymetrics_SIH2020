```
Y = df['SepsisLabel']
df.drop('SepsisLabel',inplace = True,axis=1)
X = df[df.columns]
# X = df[df.columns[:len(df.columns)-4]]

train_size = int(0.9*len(Y))
test_size = len(X)-train_size
```

```
train_size,test_size
```

```
X_train, Y_train = X[:train_size], Y[:train_size]
X_test, Y_test = X[train_size:], Y[train_size:]
```

```
# display(X_train,Y_train,X_test,Y_test)
```

```
X_train = np.array(X_train)
X_test = np.array(X_test)
Y_test = np.array(Y_test)
Y_train = np.array(Y_train)
```

```
X_train = X_train.reshape(X_train.shape[0],1,X_train.shape[1])
X_test = X_test.reshape(X_test.shape[0],1,X_test.shape[1])
Y_test = Y_test.reshape(len(Y_test),)
Y_train = Y_train.reshape(len(Y_train),)
```

```
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```
# display(X_test,X_train,Y_test,Y_train)
```

**Build Model** 

```
model = Sequential()
model.add(Bidirectional(LSTM(units = 128,batch_input_shape=(X_train.shape[0], X_train.shape[1], X_train.shape[2]))))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',metrics=['accuracy','mse'])
```

```
model.fit(X_train, Y_train, batch_size=24, epochs=30, verbose=2, shuffle=False)
```

```
y_pred = model.predict(X_test)
y_pred
```
